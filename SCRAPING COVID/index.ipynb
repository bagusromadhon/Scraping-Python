CONVERT DATE TIME 

```python
df_indonesia['key_as_string']=pd.to_datetime(df_indonesia['key_as_string'])

a=[]
tanggalnya = df_indonesia['key_as_string'].tolist()
for i in tanggalnya:
    dt = i.strftime('%Y-%m-%d')
    a.append(dt)

df_indonesia['key_as_string'] = a

df_indonesia
```

MENGCONVERT KEY AS STRING ALIAS TANGGAL DARI OBJECT KE DATE TIME

```python
df_indonesia['key_as_string']=df_indonesia['key_as_string'].astype('datetime64[ns]')
df_indonesia.info()
```

MENGGANTI NAMA COLUMNS 

```python
df_indonesia = df_indonesia.rename(columns={"key_as_string":"tanggal"})
df_indonesia.head(1)
```
